#### . Open the file using python
#### . Read every line in the file

```python
airlines_data=[]  
with open("C:\\Users\\prasu\\Downloads\\airlines.csv", "r") as f: # it opens the file in read mode
    lines=f.readlines() # to read each line in the file
    for line in lines:
        line=line.rstrip() # to remove \n at the end of eacch line
        airlines_data.append(line)
        
airlines_data      
```

```python
#getting only column names from airlines data
column_names=airlines_data[0].split(',')
```

```python
#Converting the file into dictionary format
airport_code=[]
airport_name=[]
year=[]
flight_statistics=[]
for i in range(1,len(airlines_data)):
    airport_code.append(airlines_data[i].split(',')[0])
    airport_name.append(airlines_data[i].split(',')[1]+airlines_data[i].split(',')[2])
    year.append(airlines_data[i].split(',')[3])
    flight_statistics.append(airlines_data[i].split(',')[4])
airlines_list=[]
airlines_list.append([airport_code,airport_name,year,flight_statistics])
airlines_details={}
for i in range(len(column_names)):
    airlines_details[column_names[i]]=airlines_list[0][i]
print(airlines_details)
```
